```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('datasets/office_episodes.csv')
```

```python
df.head()
```

```python
# Select only the columns of interest
df_subset = df[['episode_number','viewership_mil','guest_stars','has_guests','scaled_ratings']]
```

```python
df_subset.head(10)
```

```python
# Create a figure and increase the figure size
fig = plt.figure(figsize=(8,8))

# Create a scatter plot of duration versus year
plt.scatter(df_subset['episode_number'],df_subset['viewership_mil'])

# Create a title
plt.title("Popularity, Quality, and Guest Appearances on the Office")

# Show the plot
plt.show()
```

```python
# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset
for lab,row in df_subset.iterrows():
    if row['scaled_ratings']<=0.25 :
        colors.append('red')
    elif row['scaled_ratings']>=0.25 and row['scaled_ratings']<=0.50 : 
        colors.append('orange')
    elif row['scaled_ratings']>=0.50 and row['scaled_ratings']<=0.75 : 
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
        
# Inspect the first 10 values in your list        
colors[0:10]
```

```python
# List comprehension for maker 
markers = []

for lab,row in df_subset.iterrows():
    if row['has_guests'] == False :
        markers.append(25)
    else:
        markers.append(250)
        
markers[0:10]
```

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(df_subset['episode_number'],df_subset['viewership_mil'],s = markers,c=colors,marker='o')

# Create a title
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership(Millions)")

# Show the plot
plt.show()
```
